```python
class Producto:        
    def listar_productos():
        print('Estos son los productos que existen:\n')
        print('Nombre \t      CÃ³digo \tPrecio\tCantidad')
        try:
            path = './productos.txt'
            archivo_abierto = open(path)
            lineas_archivo = archivo_abierto.read().split(';')
            for linea in lineas_archivo:
                print(linea,'\t',end='')
                
            archivo_abierto.close()
        except:
            print('Error al listar los productos')
            
    def registrar_producto():
        nombrePro = input('Ingrese el nombre del producto:')
        codigoPro = input('Ingrese el codigo del producto:')
        precioPro = input('Ingrese el precio del producto:')
        cantidadPro = input('Ingrese la cantidad del producto:')
        try:
            path = './productos.txt'
            archivo_escritura = open(path, mode='a')
            archivo_escritura.writelines(['\n'+nombrePro+';',codigoPro+';',precioPro+';',cantidadPro])
            archivo_escritura.close()
        except:
            print('Error al momento de registrar un producto')
            
    def actualizar_producto():
        try:
            path = './productos.txt'
            nombrePro = input('Ingrese el nuevo nombre del producto:')
            codigoPro = input('Ingrese el nuevo codigo del producto:')
            precioPro = input('Ingrese el nuevo precio del producto:')
            cantidadPro = input('Ingrese la nueva cantidad del producto:')
            archivo_lectura = open(path, mode='r')
            lineas_archivo = archivo_lectura.readlines()
            productos = []
            aux = 0
            for item in lineas_archivo:
                productos = item.split(';')
                if productos[1] == str(codigoPro):
                    lineas_archivo[aux] =f'{nombrePro};{codigoPro};{precioPro};{cantidadPro}\n'
                aux = aux + 1
            archivo_lectura.close()
            archivo_escritura = open(path, mode='w')
            archivo_escritura.writelines(lineas_archivo)
            archivo_escritura.close()
        except ex:
            print('Error al momento de actualizar un producto')
            print(ex)
    
    def eliminar_producto():
        try:
            nombrePro = input('Ingrese el nombre del producto a eliminar:')
            path = './productos.txt'
            archivo_lectura = open(path, mode='r')
            lineas_archivo = archivo_lectura.readlines()
            productos = []
            aux = 0
            for item in lineas_archivo:
                aux=aux+1
                productos = item.split(';')
                if productos[0]== nombrePro:
                    lineas_archivo.pop(aux-1)
            archivo_lectura.close()
            archivo_escritura = open(path,mode='w')
            archivo_escritura.writelines(lineas_archivo)
            archivo_escritura.close()
        except:
            print('Error al eliminar el producto')
```
