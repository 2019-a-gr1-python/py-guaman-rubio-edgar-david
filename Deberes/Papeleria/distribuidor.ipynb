```python
class Distribuidor:        
    def listar_distribuidores():
        print('Estos son los distribuidores registrados:\n')
        print('Nombre \t      Dirección \t Dia visita\t N. Celular')
        try:
            path = './distribuidores.txt'
            archivo_abierto = open(path)
            lineas_archivo = archivo_abierto.read().split(';')
            for linea in lineas_archivo:
                print(linea,'\t',end='')
                
            archivo_abierto.close()
        except:
            print('Error al listar los distribuidores')
            
    def registrar_distribuidor():
        nombreDis = input('Ingrese el nombre del distribuidor:')
        direccionDis = input('Ingrese la dirección del distribuidor:')
        visitaDis = input('Ingrese el día que realiza la visita:')
        celularDis = input('Ingrese el número de celular:')
        try:
            path = './distribuidores.txt'
            archivo_escritura = open(path, mode='a')
            archivo_escritura.writelines([nombreDis+';',direccionDis+';',visitaDis+';',celularDis])
            archivo_escritura.close()
        except:
            print('Error al momento de registrar un distribuidor')
    
    def actualizar_distribuidor():
        try:
            path = './distribuidores.txt'
            nombreDis = input('Ingrese el nuevo nombre del distribuidor:')
            direccionDis = input('Ingrese la nueva dirección del distribuidor:')
            visitaDis = input('Ingrese el nuevo dia de visita del distribuidor:')
            celularDis = input('Ingrese el nuevo número de celular:')
            archivo_lectura = open(path, mode='r')
            lineas_archivo = archivo_lectura.readlines()
            productos = []
            aux = 0
            for item in lineas_archivo:
                productos = item.split(';')
                if productos[0] == nombreDis:
                    lineas_archivo[aux] =f'{nombreDis};{direccionDis};{visitaDis};{celularDis}\n'
                aux = aux + 1
            archivo_lectura.close()
            archivo_escritura = open(path, mode='w')
            archivo_escritura.writelines(lineas_archivo)
        except ex:
            print('Error al momento de actualizar un distribuidor')
            print(ex)
    
    def eliminar_distribuidor():
        try:
            nombreDis = input('Ingrese el nombre del distribuidor a eliminar:')
            path = './distribuidores.txt'
            archivo_lectura = open(path, mode='r')
            lineas_archivo = archivo_lectura.readlines()
            productos = []
            aux = 0
            for item in lineas_archivo:
                aux=aux+1
                productos = item.split(';')
                if productos[0]== nombreDis:
                    lineas_archivo.pop(aux-1)
            archivo_lectura.close()
            archivo_escritura = open(path,mode='w')
            archivo_escritura.writelines(lineas_archivo)
            archivo_escritura.close()
        except:
            print('Error al eliminar al distribuidor')
```
