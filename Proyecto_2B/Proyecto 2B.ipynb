```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot  as plt
%matplotlib inline
```

```python
path='C://Users/Edgar/Documents/GitHub/py-guaman-rubio-edgar-david/Proyecto_2B/datos/datos/Datos.csv'
```

```python
funding = pd.read_csv(path)
```

```python
funding.info()
```

```python
funding.head()
```

```python
plt.figure(figsize=(10,30))
sns.barplot(x=funding.groupby(['montoConseguido']).contribuidores.mean(),y=funding.moneda ,orient='h')
plt.ylabel("porcentaje de avance")
plt.title("Moneda más usada");
```

```python
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(20, 30), sharex=True)
sns.barplot(x=funding.contribuidores, y=funding.montoConseguido, palette="rocket", ax=ax1)
ax1.axhline(0, color="k", clip_on=False)
```

```python
fig = plt.figure(figsize=(8,18))
funding.meta[funding.contribuidores].value_counts().plot(kind='barh',alpha=1)
plt.ylabel("Monto alcanzado")
plt.show()
```

```python
plt.figure(figsize=(10,20))
sns.scatterplot(x=funding['contribuidores'], y=funding['meta'])
```

```python
plt.figure(figsize=(15,8))
funding.groupby(['contribuidores'])['porcentaje'].count().plot(kind="pie")
```

```python
funding.groupby(["porcentaje","meta"]).montoConseguido.mean().unstack().plot(kind='bar',figsize=(25,15),rot=0)
plt.ylabel('Contribuidores por proyecto');
```
