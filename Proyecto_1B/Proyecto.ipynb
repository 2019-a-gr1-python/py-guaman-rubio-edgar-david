### Seleccionamos las primeras 10 filas para tener un pequeño ejemplo de nuestro conjunto de datos.

```python
plt.figure(figsize=(20,10))
aplicaciones.groupby(['Category'])['Rating'].mean().plot(kind="bar")
```

```python
plt.figure(figsize=(20,10))
aplicaciones.groupby(['Category'])['Installs'].count().plot(kind="bar")
```

### El gráfico mostrado en la parte superior nos indica el número de descargas de acuerdo al tipo de aplicación, podemos observar que el tipo de aplicación más descargada es de tipo Familia.

```python
plt.figure(figsize=(30,10))
aplicaciones.groupby(['Genres'])['Installs'].count().plot(kind="bar")
```

```python
plt.figure(figsize=(15,8))
aplicaciones.groupby(['Type'])['Installs'].count().plot(kind="bar")
```

### A pesar de que las aplicaciones pagadas presentas mejores opciones y cero publicidad, podemos observar que las aplicaciones gratis son las más descargadas.

```python
aplicaciones.groupby(['Category'])['Content Rating'].count().plot(kind="bar")
```

### Las aplicaciones tienen un target de edad que se publica en Google Play Store. A este atributo se le conoce como Content Rating y se agrupa por Niños, Jóvenes y Adultos.

```python
plt.figure(figsize=(8,15))
sns.scatterplot(x=aplicaciones['Rating'], y=aplicaciones['Category'])
```

### Este último gráfico es interesante, ya que podemos ver un valor atípico dentro del conjunto de datos. Existe un punto que sale de toda lógica, con esto podemos ver que el data set tiene ciertos valores que no aportan mucho a la gráfica.

```python
plt.figure(figsize=(20,10))
aplicaciones.groupby(['Genres'])['Price'].count().plot(kind="bar")
```
