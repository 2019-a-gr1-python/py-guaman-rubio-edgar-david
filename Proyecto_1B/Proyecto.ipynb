### Seleccionamos las primeras 10 filas para tener un peque√±o ejemplo de nuestro conjunto de datos.

```python
plt.figure(figsize=(10,30))
aplicaciones.groupby(['Category'])['Rating'].mean().plot(kind="barh")
```

```python
plt.figure(figsize=(20,30))
aplicaciones.groupby(['Category'])['Installs'].count().plot(kind="barh")
```

### El gr√°fico mostrado en la parte superior nos indica el n√∫mero de descargas de acuerdo al tipo de aplicaci√≥n, podemos observar que el tipo de aplicaci√≥n m√°s descargada es de tipo Familia.

```python
plt.figure(figsize=(10,40))
aplicaciones.groupby(['Genres'])['Installs'].count().plot(kind="barh")
```

```python
plt.figure(figsize=(15,8))
aplicaciones.groupby(['Type'])['Installs'].count().plot(kind="pie")
```

### A pesar de que las aplicaciones pagadas presentas mejores opciones y cero publicidad, podemos observar que las aplicaciones gratis son las m√°s descargadas.

```python
plt.figure(figsize=(10,30))
aplicaciones.groupby(['Category'])['Content Rating'].count().plot(kind="barh")
```

### Las aplicaciones tienen un target de edad que se publica en Google Play Store. A este atributo se le conoce como Content Rating y se agrupa por Ni√±os, J√≥venes y Adultos.

```python
plt.figure(figsize=(10,20))
sns.scatterplot(x=aplicaciones['Rating'], y=aplicaciones['Category'])
```

### Este √∫ltimo gr√°fico es interesante, ya que podemos ver un valor at√≠pico dentro del conjunto de datos. Existe un punto que sale de toda l√≥gica, con esto podemos ver que el data set tiene ciertos valores que no aportan mucho a la gr√°fica.
